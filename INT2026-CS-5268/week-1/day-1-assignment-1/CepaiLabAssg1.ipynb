```python
# 1 check the password is weak or Strong
import re

password = input("Enter password: ")

if (len(password) >= 8 and
    re.search("[A-Z]", password) and
    re.search("[a-z]", password) and
    re.search("[0-9]", password) and
    re.search("[@#$%]", password)):
    print("Strong Password")
else:
    print("Weak Password")
```

```python
# 2 Ip Adress validator
ip = input("Enter IP Address: ")

parts = ip.split(".")

if len(parts) != 4:
    print("Invalid IPv4 Address")
else:
    valid = True
    for part in parts:
        if not part.isdigit():
            valid = False
            break
        num = int(part)
        if num < 0 or num > 255:
            valid = False
            break

    if valid:
        print("Valid IPv4 Address")
    else:
        print("Invalid IPv4 Address")
```

```python
# 3 Port Scanner
import socket

target_ip = input("Enter Target IP Address: ")

ports = [21, 22, 80, 443]

print("\nScanning Target:", target_ip)
print("-" * 30)

for port in ports:
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.settimeout(1)

    result = s.connect_ex((target_ip, port))

    if result == 0:
        print(f"Port {port} is OPEN")
    else:
        print(f"Port {port} is CLOSED")

    s.close()
```

```python
# 4 File intigrity checker (MD5/SHA256)
import hashlib

filename = input("Enter file name: ")

try:
    with open(filename, "rb") as file:
        data = file.read()

        md5_hash = hashlib.md5(data).hexdigest()
        sha256_hash = hashlib.sha256(data).hexdigest()

        print("\nMD5 Hash:", md5_hash)
        print("SHA256 Hash:", sha256_hash)

except FileNotFoundError:
    print("File not found!")
```

```python
# 5 Login Attempt Counter
correct_password = "admin123"
attempts = 3

while attempts > 0:
    password = input("Enter password: ")

    if password == correct_password:
        print("Login Successful!")
        break
    else:
        attempts -= 1
        print("Wrong password! Attempts left:", attempts)

if attempts == 0:
    print("Access Denied! Too many failed attempts.")
```

```python
#6 Brute Force Simulation
import time

real_password = "admin123"

wordlist = ["123456", "password", "admin", "admin123", "letmein"]

print("Starting Brute Force Simulation...\n")

for password in wordlist:
    print("Trying password:", password)
    time.sleep(1)

    if password == real_password:
        print("\nPassword Found!")
        print("Correct Password is:", password)
        break
else:
    print("\nPassword not found in wordlist.")
```

```python
# 7 URL Safety Checker
url = input("Enter URL: ").lower()

suspicious_words = ["login", "verify", "free"]

print("\nURL Analysis Result:")

# Check protocol
if url.startswith("https://"):
    print("Secure Protocol: HTTPS")
elif url.startswith("http://"):
    print("Insecure Protocol: HTTP")
else:
    print("Unknown Protocol")

# Check suspicious keywords
found = False
for word in suspicious_words:
    if word in url:
        print("Suspicious keyword found:", word)
        found = True

if not found:
    print("No suspicious keywords found")

# Final status
if url.startswith("https://") and not found:
    print("\nURL seems SAFE")
else:
    print("\nURL may be UNSAFE")
```

```python
# 17 Encryption & Decryption (Basic)
def encrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            base = ord('A') if char.isupper() else ord('a')
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            result += char
    return result

def decrypt(text, shift):
    return encrypt(text, -shift)

# User input
text = input("Enter text: ")
shift = int(input("Enter shift number (e.g., 3): "))

encrypted_text = encrypt(text, shift)
print("Encrypted Text:", encrypted_text)

decrypted_text = decrypt(encrypted_text, shift)
print("Decrypted Text:", decrypted_text)
```

```python
# 18 # Suspicious phishing/malicious link
url = input("Enter URL: ").lower()

suspicious_keywords = ["login", "verify", "free", "update", "secure", "account"]

found_words = []

for word in suspicious_keywords:
    if word in url:
        found_words.append(word)

if found_words:
    print("Warning! Suspicious keywords found in URL:", ", ".join(found_words))
    print("This link may be MALICIOUS or PHISHING.")
else:
    print("No suspicious keywords found. URL seems safe.")

print("\nNote: This check is based on keywords in the URL string only and does not verify if the link is live or reachable.")
```
